```python
import numpy as np
```

```python
def best_line(x, y):
    a_11 = len(x)
    a_12 = sum(x)
    a_21 = a_12
    a_22 = sum(xi**2 for xi in x)
    A = [
        [a_11, a_12],
        [a_21, a_22]
    ]
    B = [
        sum(y),
        sum(yi * xi for yi, xi in zip(y, x))
    ]
    return np.linalg.solve(A, B)
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [0, 2, 1, 3, 3, 4]
a0, a1 = best_line(x, y)
print(f'{a0=} e {a1=}')
```

```python
def p(x):
    return a0 + a1 * x
```

```python
import matplotlib.pyplot as plt

t = np.linspace(min(x), max(x), 200)
pt = [p(ti) for ti in t]

plt.scatter(x, y)
plt.plot(t, pt)
```

```python
import random

def modelo(x, a, b):
    erro = -0.5 + 1 * random.random()
    return a + b * x + erro

x = np.linspace(1, 5, 200)
a, b = 2, 1.25
y = [modelo(xi, a, b) for xi in x]
a0, a1 = best_line(x, y)
p = lambda x: a0 + a1 * x
t = np.linspace(min(x), max(x), 200)
pt = p(t)
```
